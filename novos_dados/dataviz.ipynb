```python
plt.plot( # plot faz gráfico de linha
    [2, 9, 9, 2, 2], #x
    [2, 2, 9, 9, 2]  #y
)
```

```python
plt.figure()
plt.plot( # plot faz gráfico de linha
    [2, 9, 9, 2, 2], #x
    [2, 2, 9, 9, 2]  #y
)
plt.scatter( # scatter faz pontos
    [2, 9, 9, 2, 2], #x
    [2, 2, 9, 9, 2],  #y
    label = "pontos do cubo",
    c='green',
    edgecolors= 'pink'
)

plt.title ("titulo vem aqui")
plt.xlabel("eixo x")
plt.ylabel("eixo y")
plt.legend(loc=3, bbox_to_anchor=(1, 0.5))
plt.savefig("quadrado.png")
plt.show()    
```

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0, 2 * np.pi, 1024)
data2d = np.sin(t)[:, np.newaxis] * np.cos(t)[np.newaxis, :]

fig, ax = plt.subplots()
im = ax.imshow(data2d)
ax.set_title('Pan on the colorbar to shift the color mapping\n'
             'Zoom on the colorbar to scale the color mapping')

fig.colorbar(im, ax=ax, label='Interactive colorbar')

plt.show()
```
