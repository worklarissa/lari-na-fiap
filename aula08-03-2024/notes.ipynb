```python
#informações sobre o arquivo 
df.info()
```

```python
#excluindo dados NaN e armazenando o resultado 
df_clean = df.dropna()

df_clean.info()
```

```python
#removendo dados duplicados e armazenando
df_duplicate = df_clean.drop_duplicates()

df_duplicate.info()
```

### 4. Estatisticas Descritivas

```python
plt.figure()

sns.kdeplot(df_duplicate, x ="age")

plt.show
```

```python
plt.figure(figsize=(16, 15))

sns.heatmap(
    df_duplicate.corr(),
    annot= True,
    cmap="coolwarm"
)

plt.show
```
